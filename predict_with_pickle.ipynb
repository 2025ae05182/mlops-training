# Load Pickle Model and Predict

This notebook loads the saved car price model from a pickle file and tests it with a sample input.

```python
import joblib
import pandas as pd
from pathlib import Path

model_path = Path('car_price_model.pkl')
model = joblib.load(model_path)
print(f"Loaded model from {model_path.resolve()}")
```

```python
# Sample input matching the training data columns
sample_input = pd.DataFrame([
    {
        "Make": "Toyota",
        "Model": "Camry",
        "Year": 2002,
        "Fuel_Type": "Petrol",
        "Transmission": "Automatic",
        "Engine_Size": 1.9,
        "Mileage": 32000,
        "Horsepower": 149.0,
        "Torque": 141.0,
        "Owners": 2,
        "Accident_History": 0,
        "Service_History": "Full Service",
        "Color": "Brown",
        "Body_Type": "Sedan",
        "Drivetrain": "FWD",
        "Fuel_Efficiency": 38.0,
        "Location": "CA"
    }
])

print(sample_input)
prediction = model.predict(sample_input)
print("==============================================")
print(f"Predicted selling price: {prediction[0]:,.2f}")
```

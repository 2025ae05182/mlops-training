# Load Pickle Model and Predict

This notebook loads the saved car price model from a pickle file and tests it with a sample input.

```python
import joblib
import pandas as pd
from pathlib import Path

model_path = Path('car_price_model.pkl')
model = joblib.load(model_path)
print(f"Loaded model from {model_path.resolve()}")
```

```python
# Sample input matching the training data columns
sample_input = pd.DataFrame([
    {
        "Make": "Toyota",
        "Model": "Camry",
        "Year": 2002,
        "Fuel_Type": "Petrol",
        "Transmission": "Automatic",
        "Engine_Size": 1.9,
        "Mileage": 32000,
        "Horsepower": 149.0,
        "Torque": 141.0,
        "Owners": 2,
        "Accident_History": 0,
        "Service_History": "Full Service",
        "Color": "Brown",
        "Body_Type": "Sedan",
        "Drivetrain": "FWD",
        "Fuel_Efficiency": 38.0,
        "Location": "CA"
    }
])

print(sample_input)
prediction = model.predict(sample_input)
print("==============================================")
print(f"Predicted selling price: {prediction[0]:,.2f}")
```

## Connect to the GitHub repository

Reference repository: https://github.com/SanjoyT2/BoschTrainungDataScience

This notebook can be used alongside that repository for comparison or as a starting point for further experiments.

```python
import json
import urllib.request
from urllib.error import HTTPError, URLError

repo_url = 'https://api.github.com/repos/SanjoyT2/BoschTrainungDataScience'

try:
    with urllib.request.urlopen(repo_url) as response:
        repo_data = json.load(response)

    print('Repository connected successfully!')
    print('Name:', repo_data.get('full_name'))
    print('Description:', repo_data.get('description', 'No description available'))
    print('Default branch:', repo_data.get('default_branch'))
    print('Stars:', repo_data.get('stargazers_count'))
except (HTTPError, URLError) as e:
    print(f'Unable to reach GitHub right now: {e}')
```
